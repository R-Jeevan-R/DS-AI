```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
tips.head()
```

```python
iris.head()
```

```python
titanic.head()
```

## 1. Scatter Plot (Numerical - Numerical)

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter')
#sns.scatterplot(tips,x='total_bill',y='tip')
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',size='size') #Multivariate Analysis
```

## 2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(titanic,x='pclass',y='fare',hue='sex')
```

```python
sns.catplot(titanic,x='pclass',y='age',hue='sex',kind='bar')
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.catplot(titanic,y='age',x='sex',kind='box',hue='survived')
```

## 4. Displot (Numerical - Categorical)

```python
sns.displot(titanic,x='age',kind='kde',hue='survived')
```

## 5. Heatmap (Categorical - Categorical)

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

## 6. Cluster Map (Categorical - Categorical)

```python
pd.crosstab(titanic['sibsp'],titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

## 7.PairPlot

```python
sns.pairplot(iris,hue='species')
```

## 8.Line Plot (Numerical - Numerical)

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
new  = flights.drop('month',axis=1).groupby('year').sum().reset_index()
```

```python
sns.relplot(new,x='year',y='passengers',kind='line')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year',observed=False))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year',observed=False))
```

```python
import pandas as pd , numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

```python
x = np.random.randint(1,7,size=(1000))
```

```python
s = pd.Series(x)
```

```python
s.value_counts()
```

```python
sns.catplot(data = s.value_counts(),kind='bar')
```

```python
s1 = s.value_counts()/s.value_counts().sum()
s1
```

```python
sns.catplot(s1, kind='bar')
```

```python
s3 = []
for _ in range(1000):
    a = np.random.randint(1,7)
    b = np.random.randint(1,7)
    s3.append(a + b)
s3 = pd.Series(s3)
```

```python
s3.value_counts().sort_values()
```

```python
s4 = s3.value_counts() / s3.value_counts().sum()
```

```python
sns.catplot(s4,kind='bar')
```

```python
np.cumsum(s4).plot(kind='bar')
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.Series([1,1,2,3,1,3,4,2,4,5,1,8,5,4,3,2,5,2,3,1])
```

```python
x.sample()
```

```python
mean = x.mean()
```

```python
x.value_counts()
```

```python
mode = x.mode()[0]
```

```python
median= x.median()
```

```python
sns.displot(x,kind='hist',kde = True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median:.2f}')
plt.axvline(mode, color='yellow',linestyle='dashed',linewidth=1,label=f'Mode: {mode:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution with Mean and Median')
plt.legend()

# Show the plot
plt.show()
```

```python
def Mean(series):
    Sum = 0
    for value in series:
        Sum += value
    return Sum/len(series)
```

```python
def Median(series):
    series = series.sort_values(ignore_index = True)
    l = len(series)
    if l != 0:
        return series[l//2] * 1.0
    else:
        return (series[l//2 - 1] + series[l//2])/2
    
```

```python
from collections import defaultdict
def Mode(series):
    freq = defaultdict(int)
    for value in series:
        freq[value] += 1
    Max = float('-inf')
    Max_value = 0
    for key, value in freq.items():
        if value > Max_value:
            Max_value = value
            Max = key
    return Max
```

```python
print(Mean(x))
print(Median(x))
print(Mode(x))
```

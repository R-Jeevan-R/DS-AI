```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,20)
```

```python
plt.plot(x,x,'b-',x,x**2,'g^',x,x**3,'ro')
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.show()
```

```python
df = sns.load_dataset('iris')
```

```python
df.info()
```

```python
df.shape
```

```python
df.value_counts('species')
```

```python
plt.figure(figsize=(9,3))
plt.subplot(131)
bar = plt.bar(df['species'].unique(),df['species'].value_counts())
plt.setp(bar,color='k',edgecolor='r')
plt.subplot(132)
scatter = plt.scatter(df['species'].unique(),df['species'].value_counts())
plt.setp(scatter,color='y')
plt.subplot(133)
line = plt.plot(df['species'].unique(),df['species'].value_counts())
plt.setp(line,linewidth=3.0,color='g',marker='+')
```
